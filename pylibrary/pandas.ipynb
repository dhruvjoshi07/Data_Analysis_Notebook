```python
import numpy as np
import pandas as pd
```

```python
data={
    'Category':['A','B','A','B','A','B','A','B'],
    'Store':['S1','S1','S2','S2','S1','S2','S2','S1'],
    'Sales':[100,200,150,250,120,180,200,300],
    'Quantity':[10,15,12,18,8,20,15,25],
    'Date':pd.date_range('2023-01-01',periods=8)
}
df=pd.DataFrame(data,index=[1,2,3,4,5,6,7,8])
df
```

```python
#group by category and calculate the sum of sales
cat=df.groupby('Category')
cat
for i,v in cat:
    print(i)
    print(v)
```

```python
cat=df.groupby(['Category'])['Sales'].sum()
cat
```

```python
#group by multiple columns
# group by category and store
cat=df.groupby(['Category','Store'])['Sales'].sum()
cat
```

```python
#Aggregation
df['Sales'].mean()
```

```python
df['Sales'].agg(['count','max'])       #only mode can't do
```

```python
#PivotTables
import numpy as np
import pandas as pd
```

```python
data = {
    'Date': pd.date_range('2023-01-01', periods=20),
    'Product': ['A', 'B', 'C', 'D'] * 5,
    'Region': ['East', 'West', 'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    'Sales': np.random.randint(100, 1000, 20),
    'Units': np.random.randint(10, 100, 20),
    'Rep': ['John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice']
}
df=pd.DataFrame(data)

df['Month']=df['Date'].dt.month_name()
df['Quarter']='Q'+df['Date'].dt.quarter.astype(str)
df
```

```python
#value ka mean aa ra
pd.pivot_table(df,values="Sales",index='Region',columns="Product")
```

```python
pd.pivot_table(df,values="Sales",index='Region',columns="Product",fill_value="-",aggfunc="median")
```

```python
#values whi lo jo numeric ho but index can different
pivot2=pd.pivot_table(df,values=["Sales","Units"],index="Region",columns="Product",fill_value="-")
pivot2
```

```python
#cross tab(yaha ham counting fx use krte rather than aggreagting fx)
pd.crosstab(df['Region'],df['Product'])     #counting function
```
