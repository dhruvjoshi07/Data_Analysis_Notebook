```python
import numpy as np
import pandas as pd
```

```python
#nan mtln nonevalue
data={
    'A':[1,2,np.nan,4,5],
    'B':[np.nan,2,3,4,5],
    'C':[1,2,3,np.nan,np.nan],
    'D':[1,np.nan,np.nan,np.nan,5]
}
df=pd.DataFrame(data)
df
```

```python
#ham true false mei dekh lenge ki her rows mei nan value hai ki nhi using 'isna'
df.isna()
```

```python
df.isna().sum()
```

```python
#kya sb mei null vale hau?
df.isna().any()
```

```python
'''Removing works on rows'''
df.dropna()   #jis jsi mei null value hai hata do woh row
```

```python
#thresh
df.dropna(thresh=3)    #atleast 3 none null val honi chahiye
```

```python
#filling the non value data
```

```python
df
```

```python
df.fillna(0)    #jaha nan hai waha 0
```

```python
values={'A':1,'B':2,'C':3,'D':4}
df.fillna(value=values)
```

```python
df
```

```python
df.fillna(df.mean())
```
