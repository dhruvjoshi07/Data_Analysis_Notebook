We use distribution plots to visualize the distribution of quantitative data
Lets discuss some plots that allow us to viualize the  distribution of a data set.These plots are:
distplot/histplot,
joinplot,
pairplot,
rugplot

```python
# age is quantativeive and gender is qualative
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

```python
#histogram
sns.displot(df['total_bill'])
```

```python
sns.histplot(df['total_bill'],bins=30,kde=True)  #kernal density estimater shows data skewness
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins=20,kde=True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins=20,kde=True)
```

```python
#jointplot
sns.jointplot(x='total_bill',y='tip',data=df)
```

```python
# jointplot
sns.jointplot(x='total_bill',y='tip',data=df,kind="kde")
```

```python
# jointplot
sns.jointplot(x='total_bill',y='tip',data=df,kind='hex')
```

```python
# jointplot
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg')
```
